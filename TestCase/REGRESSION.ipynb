### IMPORT

```python
import matplotlib.pyplot as plt
import numpy as np
from CLibrary import create_linear_model, fit_regression, useTrainModel, toArray
```

### LINEAR SIMPLE MODEL X2

#### INITIALISATION

```python
X = np.array([
      [1],
      [2]
])
Y = np.array([
      2,
      3
])
```

```python
plt.scatter(X,Y)
plt.show()
plt.clf()
```

#### ENTRAINEMENT

```python
W = create_linear_model(2)
fit_regression(W, toArray(X, 2, 1), 2, 1, toArray(Y, 1, 2))
```

#### UTILISATION

```python
abscisse = np.array([[0], [1], [2], [3]])
ordonne = []
for abs in abscisse:
    ordonne.append(useTrainModel(abs))
    
# points initiaux
plt.scatter(X,Y)

plt.plot(abscisse, ordonne)
plt.show()
```

## LINEAR SIMPLE 3D

#### INITIALISATION

```python
X = np.array([
      [1, 1],
      [2, 2],
      [3, 1]
])
Y = np.array([
      2,
      3,
      2.5
])
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],Y)
plt.show()
plt.clf()
```

#### ENTRAINEMENT

```python
W = create_linear_model(3)
fit_regression(W, toArray(X, 3, 2), 3, 2, toArray(Y, 1, 3))
```

#### UTILISATION

```python
Xtest = np.array([[a / 10, b / 10] for a in range(30) for b in range(30)])
Ytest = np.array([[useTrainModel(c)] for c in Xtest])
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],Y)

#modèle
ax.scatter(Xtest[:,0],Xtest[:,1],Ytest)

plt.show()
plt.clf()
```
