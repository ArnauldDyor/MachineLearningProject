```python
# CE NOTEBOOK N'EST QU'UNE PREMICE DE NOS VERITABLES TEST CASES, JE N'AI PAS IMPORTE NOS FONCTIONS, UNIQUEMENT LES RESULTATS
# Cf le script .py pour obtenir les resultats

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
#Regression linéaire cas n ° 1
#sortie après entrainement [1, 1]
X = np.array([
      [1],
      [2]
])
Y = np.array([
      2,
      3
])

W = [1, 1]
```

```python
XTrain = np.array([[a]for a in range(5)])
YTrain = np.array([W[0] * a + W[1] for a in range(5)])

plt.scatter(X, Y)
plt.plot(XTrain, YTrain)
plt.show()
```

```python
# Regression linéaire cas n°2
# sortie de l'entrainement [0.25, 2]

X = np.array([
      [1],
      [2],
      [3]
])
Y = np.array([
      2,
      3,
      2.5
])

W = [0.25, 2]
```

```python
XTrain = np.array([[a]for a in range(5)])
YTrain = np.array([W[0] * a + W[1] for a in range(5)])

plt.scatter(X, Y)
plt.plot(XTrain, YTrain)
plt.show()
```

```python
# Regression linéaire : cas n ° 3
# sortie de l'entrainement [0.5, -1, 3.5]

X = np.array([
      [1, 1],
      [2, 2],
      [3, 1]
])
Y = np.array([
      2,
      3,
      2.5
])

W = [0.5, -1, 3.5]
```

```python
XTrain = np.array([[a, b]for a in range(10) for b in range(10)])
YTrain = np.array([W[0] * a + W[1] * b + W[2] for a in range(10) for b in range(10)])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],Y)
ax.scatter(XTrain[:,0],XTrain[:,1],YTrain)
plt.show()
plt.clf()
```
