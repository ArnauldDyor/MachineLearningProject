### IMPORTS

```python
import matplotlib.pyplot as plt
import numpy as np
from CLibrary import create_linear_model, fit_classification_rosenblatt_rule, useTrainModel, toArray
```

### LINEAR SIMPLE

#### INITIALISATION

```python
X = np.array([
      [1, 2],
      [2, 3],
      [3, 3]
])
Y = np.array([
      1,
      -1,
      -1
])
```

```python
plt.scatter(X[0, 0], X[0, 1], color='blue')
plt.scatter(X[1:3,0], X[1:3,1], color='red')
plt.show()
plt.clf()
```

#### ENTRAINEMENT

```python
W = create_linear_model(2)
fit_classification_rosenblatt_rule(W, toArray(X, 3, 2), 3, 2, 1, toArray(Y, 1, 3), 0.1, 10000)
```

#### UTILISATION

```python
SEP = np.array([[a / 10, b / 10]for a in range(10, 30) for b in range(10, 30)])
for case in SEP:
    res = useTrainModel(case)
    if(res < 0):
        plt.scatter(case[0], case[1], color='red')
    else : 
        plt.scatter(case[0], case[1], color='blue')

plt.show()
plt.clf()
```

### CROSS

#### INITIALISATION

```python
X = np.random.random((500, 2)) * 2.0 - 1.0
Y = np.array([1 if abs(p[0]) <= 0.3 or abs(p[1]) <= 0.3 else -1 for p in X])
```

```python
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]] == 1, enumerate(X)))))[:,1], color='blue')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]] == -1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]] == -1, enumerate(X)))))[:,1], color='red')
plt.show()
plt.clf()
```

#### ENTRAINEMENT

```python
W = create_linear_model(2)
fit_classification_rosenblatt_rule(W, toArray(X, 500, 2), 500, 2, 1, toArray(Y, 1, 500), 0.01, 100)
```

#### UTILISATION

```python
SEP = np.array([[a / 10, b / 10 ]for a in range(-10, 10) for b in range(-10, 10)])
for case in SEP:
    res = useTrainModel(case)
    if(res < 0):
        plt.scatter(case[0], case[1], color='blue')
    else : 
        plt.scatter(case[0], case[1], color='red')

plt.show()
plt.clf()
```

### MULTI LINEAR 3 CLASSES

#### INITIALISATION

```python
X = np.random.random((500, 2)) * 2.0 - 1.0
Y = np.array([[1, 0, 0] if -p[0] - p[1] - 0.5 > 0 and p[1] < 0 and p[0] - p[1] - 0.5 < 0 else 
              [0, 1, 0] if -p[0] - p[1] - 0.5 < 0 and p[1] > 0 and p[0] - p[1] - 0.5 < 0 else 
              [0, 0, 1] if -p[0] - p[1] - 0.5 < 0 and p[1] < 0 and p[0] - p[1] - 0.5 > 0 else 
              [0, 0, 0]for p in X])
```

```python
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,1], color='blue')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,1], color='red')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,1], color='green')
plt.show()
plt.clf()
```

#### ENTRAINEMENT

```python
W = create_linear_model(2)
fit_classification_rosenblatt_rule(W, toArray(X, 500, 2), 500, 2, 3, toArray(Y, 500, 3), 0.01, 100)
```

#### UTILISATION

```python
SEP = np.array([[a / 10, b / 10 ]for a in range(-10, 10) for b in range(-10, 10)])
for case in SEP:
    res = useTrainModel(case)
    if(res == 0):
        plt.scatter(case[0], case[1], color='blue')
    elif(res == 1) : 
        plt.scatter(case[0], case[1], color='red')
    elif(res == 2):
        plt.scatter(case[0], case[1], color='green')
 
plt.show()
plt.clf()
```
